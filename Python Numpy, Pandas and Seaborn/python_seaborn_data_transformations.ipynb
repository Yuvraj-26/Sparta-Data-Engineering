```python
import numpy as np
import pandas as pd
# Seaborn is a Python data visualization library based on matplotlib. 
# It provides a high-level interface for drawing 
# informative statistical graphics.
import seaborn as sns
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
# returns descriptive statistics
#that summarize the central tendency, dispersion and shape 
#of a dataset's distribution, excluding NaN values
iris.describe()
```

```python
# Pandas groupby is used for grouping the data according to the categories 
# and applying a function to the categories.

# dataframe.groupby() function is used to split the data into groups 
# based on some criteria
spec = iris.groupby('species')
spec.head()
```

```python
spec.describe()
```

```python
# filtering a dataset called "iris" to create three separate dataframes, 
# one for each species of iris flowers: 'versicolor', 'setosa', and 'virginica'.
vers = iris[iris['species'] == 'versicolor']
setosa = iris[iris['species'] == 'setosa']
virg = iris[iris['species'] == 'virginica']
virg
```

```python
# creating two new DataFrames: "sepal_info" and "petal_info" 
# by selecting specific columns from the original "iris" DataFrame.
sepal_info = pd.DataFrame({'sepal_length':iris['sepal_length'],'sepal_width':iris['sepal_width'],'species':iris['species']})
petal_info = pd.DataFrame({'petal_length':iris['petal_length'],'petal_width':iris['petal_width'],'species':iris['species']})
sepal_info
```

```python
# merge
# inner join (intersection) between two DataFrames, 
# "sepal_info" and "petal_info," using their index values 
#  as the key for merging.
pd.merge(sepal_info, petal_info, how='inner', right_index=True, left_index=True)
```

```python
# Using the pd.concat function to concatenate two DataFrames, 
# vers and setosa, along the vertical axis 
# axis=0: Vertical axis, used for operations along rows.
# axis=1: Horizontal axis, used for operations along columns.
pd.concat([vers, setosa], axis=0)
```

```python
# pivot and stack (unpivot) needed for transforming long or wide format
# data for creating charts such as heatmaps
# flights dataset is in long format
flights = sns.load_dataset("flights")
flights
```

```python
# when creating heatmap, function expects data to be organised
# in wide format to create colours

# pivoting
f_piv = flights.pivot(index='year',columns='month',values='passengers')
f_piv
```

```python
# stacking
# It stacks (reshapes) the 'f_piv' DataFrame, which is a pivot table, 
# into a Series with a multi-level index. This essentially transforms 
# the wide-format data into a long-format data structure.
f_stack = f_piv.stack()
f_stack
type(f_stack)
```

```python
# unstack on series level
f_stack.unstack()
```
